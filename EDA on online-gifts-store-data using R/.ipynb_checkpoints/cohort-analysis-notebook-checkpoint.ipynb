```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('cohort_data_GER.csv')
```

```python
data.head()
data.drop('Unnamed: 0', axis= 1, inplace= True)
```

```python
data.head()
```

```python
data.loc[:, 'Invoice Date'] = pd.to_datetime(data['Invoice Date'])
```

```python
data.info()
```

```python
# assign aquisition month cohort
def get_month(x):
    return dt.datetime(x.year, x.month, 1)

data['Invoice Month'] = data['Invoice Date'].apply(get_month)

data['cohort month']= data.groupby('Customer ID')['Invoice Month'].\
                                 transform('min')
```

```python
data.head()
```

```python
data.head()
```

```python
data.head()
```

```python
# extract integer values from the data

def get_date_int(df, col):
    year= df[col].dt.year
    month = df[col].dt.month
    day = df[col].dt.day
    return year, month, day
```

```python
# assign time offset value

invoice_year, invoice_month, _= get_date_int(df= data,  col= 'Invoice Month')
cohort_year, cohort_month, _ = get_date_int(data,  'cohort month')

year_diff= invoice_year - cohort_year
month_diff= invoice_month - cohort_month
```

```python
data['cohort index'] = year_diff + 12+month_diff+1
```

```python
data.head()
```

```python
data['cohort index'].unique()
```

```python
# monthly active customers for each cohort

cohort_data= data.groupby(['cohort month', 'cohort index'])['Customer ID'].\
                               apply(pd.Series.nunique).\
                                reset_index()
```
