# bank account availablity to individuals in Africa

**objective**<br>
Although there is numerous availabilty of mobile banking, Bank account is still needed for legal reasons, credit scoring etc.
The aim of the project is to discover the prior reason many people in certain african country do not have a bank account.
This could help the banking firm target specific audience in thier ad-placement and always make changes in the way the operate.

**data**<br>
The data used for this analysis contains thousands of rows of information of individuals such as the country they live, accessibility to cellphone, job type, age, household size, education level etc.

**expectation**
* people with less education level is likely to not have a bank account.
* people of the age less than 18 and greater than 70 is not likely to have a bank account.
* unemployed people is less likely to have a bank account
etc.

**data analysis process**
* check the packaging
* check for counts
* check for distribution
* check for relations
* make pivot tables

```python
# chebychevs' rule
mean_age= data['Age of respondent'].mean()
std_age= data['Age of respondent'].std()

print('1. 75% of individuals have ages between ', np.floor(mean_age-2*std_age), ' and ', 
      np.floor(mean_age+2*std_age))
print('2. 88.89% of individuals have ages between ', np.floor(mean_age-2*std_age), ' and ', 
      np.floor(mean_age+3*std_age))
```

```python
pd.DataFrame(data['Age of respondent'].describe([.25, .5, .75, .95])).round(2)
```

```python
# confidence interval of ages (of population mean)

mu = data['Age of respondent'].mean()
sd_age = data['Age of respondent'].std()
se = sd_age/np.sqrt(data['Age of respondent'].count())
ci = mu-1.96*se, mu+1.96*se
print('Confidence Interval of mean ages id between ',np.floor(ci)[0], ' and ', np.floor(ci)[1] )
```
