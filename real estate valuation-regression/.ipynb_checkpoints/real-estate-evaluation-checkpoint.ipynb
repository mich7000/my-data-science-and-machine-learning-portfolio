# data cleaning & fetaure engineering

```python
columns = ['No.','transaction date', 'house age', 'dist to nearest MRT', 'number of convenience stores',
          'latitude', 'longitude', 'house price of unit area']
data.columns = columns
```

```python
data['transaction date'] = np.round(data['transaction date'], 3)
data.drop('No.', axis= 1, inplace=True)
```

```python
data.head()
```

```python
# change the decimal years to real dates
from datetime import datetime, timedelta
def convert(number):
    
    year= int(number)
    d = timedelta(days= (number - year)*365)
    day1 = datetime(year, 1, 1)
    date = d+day1
    return date

data['real date'] = [convert(i) for i in data['transaction date'] ]
```

```python
data.head()
```

```python
data['transaction_year'] = data['real date'].dt.strftime('%Y')
data['transaction_month'] = data['real date'].dt.strftime('%B')
data['transaction_dayname'] = data['real date'].dt.strftime('%A')
data['transaction_day'] = data['real date'].dt.strftime('%d')
```

```python
data.head()
```

# Model building phase

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import mean_absolute_error, r2_score

from sklearn.linear_model import RandomizedLasso, LinearRegression
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# fitting the model with the outliers included
cols = [ 'transaction_year', 'transaction_month', 'transaction_day','transaction_dayname',
         'house age', 'dist to nearest MRT', 'number of convenience stores',
         'latitude', 'longitude']

X, y = data[cols], data['house price of unit area']
X_train, X_test, y_train, y_test = train_test_split(X, y)
cat_cols = ['transaction_month', 'transaction_day', 'number of convenience stores']
num_cols = ['transaction_year', 'house age', 'dist to nearest MRT', 'latitude', 'longitude']

bag = BaggingRegressor(base_estimator= RandomForestRegressor(random_state= 50), 
                       n_estimators=60)


ohe = OneHotEncoder()
sc = StandardScaler()
col_trans = ColumnTransformer([('encoder', ohe, cat_cols), 
                               ('scaler',sc, num_cols)])
model_pipe= Pipeline([('transformer', col_trans), ('model', bag)])
model_pipe.fit(X_train, y_train)

y_pred = model_pipe.predict(X_test)
print('R2-score: ', np.round(r2_score(y_test, y_pred), 3))
print('mean abs error: ', np.round(mean_absolute_error(y_test, y_pred), 3))

plt.figure(figsize= (12, 7))
plt.plot(range(1, len(X_test)+1), y_test, label= 'true')
plt.plot(range(1, len(X_test)+1), y_pred, color ='red', label= 'predicted')
plt.title('True value vs Predcited', size=  20)
plt.legend()
```

```python
# fitting the model with the outliers removed
cols = [ 'transaction_year', 'transaction_month', 'transaction_day','transaction_dayname',
         'house age', 'dist to nearest MRT', 'number of convenience stores',
         'latitude', 'longitude']

X1, y1 = data2[cols], data2['house price of unit area']
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1)
cat_cols = ['transaction_month', 'transaction_day', 'number of convenience stores']
num_cols = ['transaction_year', 'house age', 'dist to nearest MRT', 'latitude', 'longitude']

bag = BaggingRegressor(base_estimator= RandomForestRegressor(random_state= 40), 
                       n_estimators=60)


ohe = OneHotEncoder()
sc = StandardScaler()
col_trans = ColumnTransformer([('encoder', ohe, cat_cols), 
                               ('scaler',sc, num_cols)])
model_pipe1= Pipeline([('transformer', col_trans), ('model', bag)])
model_pipe1.fit(X_train1, y_train1)

y_pred1= model_pipe1.predict(X_test1)
print('R2-score: ', np.round(r2_score(y_test1, y_pred1), 3))
print('mean abs error: ', np.round(mean_absolute_error(y_test1, y_pred1), 3))

plt.figure(figsize= (12, 7))
plt.plot(range(1, len(X_test1)+1), y_test1, label= 'true')
plt.plot(range(1, len(X_test1)+1), y_pred1, color ='red', label= 'predicted')
plt.title('True value vs Predcited', size=  20)
plt.legend()
```
