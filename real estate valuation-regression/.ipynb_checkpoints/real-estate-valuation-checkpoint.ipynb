```python
columns = ['No.','transaction date', 'house age', 'dist to nearest MRT', 'number of convenience stores',
          'latitude', 'longitude', 'house price of unit area']
df.columns = columns
```

```python
df['transaction date'] = np.round(df['transaction date'], 3)
```

```python
# change the decimal years to standard years...

from datetime import datetime, timedelta
def convert(number):
    
    year= int(number)
    d = timedelta(days= (number - year)*365)
    day1 = datetime(year, 1, 1)
    date = d+day1
    return date

df['date-time'] = [convert(i) for i in df['transaction date'] ]
```

```python
#df.drop('transaction date', axis=1, inplace=True)
```

```python
# creates new variables of date components
df['transaction_year'] = df['date-time'].dt.strftime('%Y')
df['transaction_month'] = df['date-time'].dt.strftime('%B')
df['transaction_dayname'] = df['date-time'].dt.strftime('%A')
df['transaction_day'] = df['date-time'].dt.strftime('%d')
```

```python
df.drop('transaction date', axis=1, inplace=True)
```

# BUILDING THE MODEL

```python
# all necessary packages to aid in the model building processes
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error # root mean square error

from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
df.head()
```

```python
df.drop(['date-time'], axis=1, inplace=True)
```

```python
list(df)
```

## OBJECT INITIALIZATION AND DATA SPLITTING

```python
cat = ['transaction_month', 'transaction_day', 'number of convenience stores']
num = ['transaction_year', 'house age', 'dist to nearest MRT', 'latitude', 'longitude']
ohe = OneHotEncoder()
sc = StandardScaler()
col_trans = ColumnTransformer([('encoder', ohe, cat), 
                               ('scaler',sc, num)])

cols = ['house age',
 'dist to nearest MRT',
 'number of convenience stores',
 'latitude',
 'longitude',
 'transaction_year',
 'transaction_month',
 'transaction_dayname',
 'transaction_day']

X, y = df[cols], df['house price of unit area']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=45, test_size=.3)
```

```python
# various algorithms
lm= LinearRegression()
rfr = RandomForestRegressor(max_depth=50, max_features=12, max_samples=4)
lassr = Lasso()
dtr= DecisionTreeRegressor(max_features=20, max_depth=100)
svr = SVR()
```

# MODEL FITTING AND EVALUATION

```python
#########################
#LINEAR REGRESSION MODEL
########################

model1= Pipeline(steps=[('column_transformer', col_trans), ('classifier', lm)])
model1.fit(X_train, y_train);

# prediction andn evaluation
ypred1 = model1.predict(X_test)
predictions= pd.DataFrame({'Actual': y_test, 'Predicted':ypred1})

# measures
lm_r2= r2_score(y_test, ypred1)
lm_mse = mean_squared_error(y_test, ypred1)
lm_mae = mean_absolute_error(y_test, ypred1)
lm_rmse = np.sqrt(lm_mse)
report1= pd.Series({'R2-score':lm_r2, 
                      'Mean Square Error':lm_mse, 
                      'Mean Absolute Error': lm_mae, 
                      'Root Mean Square Error': lm_rmse})
```

```python
report1
```

```python
###########
# RANDOM FOREST REGRESSOR
###########
model2= Pipeline(steps=[('column_transformer', col_trans), ('classifier', rfr)])
model2.fit(X_train, y_train);

# prediction andn evaluation
ypred2 = model2.predict(X_test)
predictions2= pd.DataFrame({'Actual': y_test, 'Predicted':ypred1})

# measures
rfr_r2= r2_score(y_test, ypred1)
rfr_mse = mean_squared_error(y_test, ypred2)
rfr_mae = mean_absolute_error(y_test, ypred2)
rfr_rmse = np.sqrt(rfr_mse)
report2= pd.Series({'R2-score':rfr_r2, 
                      'Mean Square Error':rfr_mse, 
                      'Mean Absolute Error': rfr_mae, 
                      'Root Mean Square Error': rfr_rmse})
```

```python
report2
```

```python
#########################
# DECISION TREE
####################

model3= Pipeline(steps=[('column_transformer', col_trans), ('classifier', dtr)])
model3.fit(X_train, y_train);

# prediction andn evaluation
ypred3 = model3.predict(X_test)
predictions3= pd.DataFrame({'Actual': y_test, 'Predicted':ypred3})

# measures
dtr_r2= r2_score(y_test, ypred3)
dtr_mse = mean_squared_error(y_test, ypred3)
dtr_mae = mean_absolute_error(y_test, ypred3)
dtr_rmse = np.sqrt(dtr_mse)
report3= pd.Series({'R2-score':dtr_r2, 
                      'Mean Square Error':dtr_mse, 
                      'Mean Absolute Error': dtr_mae, 
                      'Root Mean Square Error': dtr_rmse})
```

R2 score of all the models are around .5, this may improve including other measures ones we deal with the outliers in the data.
But removing those extreme values comes with penalty because they are not really anomalies they are just higher than average.
The best thing to do is to build two models (one model build with that extreme values in the data and the other model build without the extreme values).

```python
k = KFold(n_splits=20)
```

```python
models = [model1, model2, model3, model4, model5, model6]
model_name = ['linear regression', 'random forest', 'decision tree', 'SVR', 'lasso', 'ridge']

mae_= []
r2_= []
for model in models:
        mae= cross_val_score(model, X_train, y_train, cv=k, scoring='neg_mean_absolute_error')
        mae_.append(mae)
        
        r2= cross_val_score(model, X_train, y_train, cv=k, scoring='r2')
        r2_.append(r2)         
```

```python
cv_scores_mae = pd.DataFrame(mae_, index=model_name).T
cv_scores_r2 = pd.DataFrame(r2_, index=model_name).T
```

```python
cv_scores_mae.plot(kind= 'box', figsize=(10, 7), title='negative MAE')
```

```python
cv_scores_r2.plot(kind='box', figsize=(10, 7), title='R-Square')
```

```python
# we re-train decision tree regressor and  (using gridsearch)

decision_pipe = Pipeline(steps=[('transformer', col_trans), ('reg', DecisionTreeRegressor(max_features=20, max_depth=100))])
param_grid = {
    'reg__max_features': [5, 10, 15, 20, 25],
    'reg__max_depth': [100, 200, 300, 400, 500],
}
search0 = GridSearchCV(decision_pipe, param_grid, n_jobs=-1)
search0.fit(X_train, y_train);
```

```python
search0.best_estimator_
```

```python
search_pred = search0.predict(X_test)
pd.DataFrame({"actual":y_test, "predicted":search_pred})
```

```python
plt.figure(figsize=(15, 4))
plt.title('Decision tree Regressor Prediction (After Grid Search)')
plt.plot(range(1, len(X_test)+1), y_test, label= 'true')
plt.plot(range(1, len(X_test)+1), search_pred, color ='red', label= 'predicted')
plt.legend()
```
